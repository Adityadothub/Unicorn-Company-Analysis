```python
# STEP 1: Install and Import Libraries
import sqlite3
import pandas as pd

# STEP 2: Create SQLite Database in Memory
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# STEP 3: Create Unicorn Companies Table
cursor.execute('''
CREATE TABLE unicorns (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    company_name TEXT,
    valuation REAL,
    industry TEXT,
    country TEXT,
    year_founded INTEGER,
    date_joined TEXT
)
''')

# STEP 4: Insert Sample Data
sample_data = [
    ("ByteDance", 140.0, "Artificial Intelligence", "China", 2012, "2017-04-01"),
    ("SpaceX", 127.0, "Aerospace", "USA", 2002, "2012-10-01"),
    ("Stripe", 95.0, "Fintech", "USA", 2010, "2014-01-01"),
    ("Klarna", 45.6, "Fintech", "Sweden", 2005, "2011-06-01"),
    ("Databricks", 43.0, "Big Data", "USA", 2013, "2019-02-01"),
    ("Byju's", 22.0, "Edtech", "India", 2011, "2017-08-01"),
    ("Rivian", 27.6, "Automotive", "USA", 2009, "2019-09-01"),
    ("Ola Cabs", 7.0, "Transportation", "India", 2010, "2015-02-01"),
]

cursor.executemany('''
INSERT INTO unicorns (company_name, valuation, industry, country, year_founded, date_joined)
VALUES (?, ?, ?, ?, ?, ?)
''', sample_data)

conn.commit()

print("✅ Database and Table Created with Sample Data!")
```

```python
df = pd.read_sql_query("SELECT * FROM unicorns", conn)
df
```

```python
# @title company_name

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('company_name').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
top_5 = pd.read_sql_query('''
SELECT company_name, valuation
FROM unicorns
ORDER BY valuation DESC
LIMIT 5
''', conn)
top_5
```

```python
country_count = pd.read_sql_query('''
SELECT country, COUNT(*) as total_unicorns
FROM unicorns
GROUP BY country
ORDER BY total_unicorns DESC
''', conn)
country_count
```

```python
industry_valuation = pd.read_sql_query('''
SELECT industry, AVG(valuation) as avg_valuation
FROM unicorns
GROUP BY industry
ORDER BY avg_valuation DESC
''', conn)
industry_valuation
```

```python
year_founded_count = pd.read_sql_query('''
SELECT year_founded, COUNT(*) as companies_founded
FROM unicorns
GROUP BY year_founded
ORDER BY year_founded ASC
''', conn)
year_founded_count
```
